##### Test with their Hypothesis


1. The Shapiro-walk test is the test of checking normality
    - Null Hypothesis(H0) : Data are normally distributed
    - Alternative Hypothesis(H1) : Data are not normally distributed


2. The Durbin Watson Test is a measure of autocorrelation (also called serial correlation) in residuals from regression analysis.
    - Null Hypothesis(H0) : no first order autocorrelation
    - Alternate Hypothesis(H1) : first order correlation exists


3. The Breusch-Pagan Test is a test for heteroscedasticity of errors in regression
    - Null Hypothesis(H0) : error variances are all equal.
    - Alternate Hypothesis(H1) : error variances are not equal. More specifically, as Y increases, the variances increase (or decrease).
    
    
4. The Jarque-Bera Test,a type of Lagrange multiplier test, is a test for normality.
    - Null Hypothesis(H0) : Data are normally distributed
    - Alternative Hypothesis(H1) : Data are not normally distributed
    
    
5. The Anderson-Darling Goodness of Fit Test (AD-Test) is a measure of how well your data fits a specified distribution. It’s commonly used as a test for normality.
    - Null Hypothesis(H0) : The data comes from a specified distribution.
    - Alternative Hypothesis(H1) : The data does not come from a specified distribution.
